##### 1. Find the Top 10 artists by Likes ([group by](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html) Artist and sum the Likes column)
Note: The Likes column has null values, so you'll need to decide how to handle those.

Other references that may be helpful include:
* [sort_values](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sort_values.html) 
* [nlargest/nsmallest](https://datascientyst.com/get-top-10-highest-lowest-values-pandas/)

```python
# how many null values do you need to handle?
# how does this influence how you'll proceed?

print("Likes column NaN/Nulls: ", df["Likes"].isna().sum())
print("Total df records: ", len(df.index))
print(
    "Percent NaN/Null: ",
    round((df["Likes"].isna().sum() / len(df.index)) * 100, 2),
    "%",
)
```

```python
# handle nulls
# you can choose to create a df copy or to edit the df directly
# if you want to play with different ways of handling nulls, create a copy

# here are examples of how to use each method we went over: drop and fillna (mean, median, mode)

# drop
df_drop = df.copy()
df_drop.dropna(subset=["Likes"], inplace=True)
print("DROP METHOD")
print("Likes column NaN/Nulls after drop: ", df_drop["Likes"].isna().sum())
print("Total df records original: ", len(df.index))
print("Total df records after drop: ", len(df_drop.index))

# average
df_avg = df.copy()
df_avg["Likes"].fillna((df_avg["Likes"].mean()), inplace=True)
print("\nFILLNA AVERAGE")
print("Likes column NaN/Nulls after fillna average: ", df_avg["Likes"].isna().sum())
print("Total df records original: ", len(df.index))
print("Total df records after fillna average: ", len(df_avg.index))

# median
df_med = df.copy()
df_med["Likes"].fillna((df_med["Likes"].median()), inplace=True)
print("\nFILLNA MEDIAN")
print("Likes column NaN/Nulls after fillna median: ", df_med["Likes"].isna().sum())
print("Total df records original: ", len(df.index))
print("Total df records after fillna median: ", len(df_med.index))

# mode
df_mod = df.copy()
df_mod["Likes"].fillna((df_mod["Likes"].mode()[0]), inplace=True)
print("\nFILLNA MODE")
print("Likes column NaN/Nulls after fillna mode: ", df_mod["Likes"].isna().sum())
print("Total df records original: ", len(df.index))
print("Total df records after fillna mode: ", len(df_mod.index))
```

```python
# group likes by artist

# selected drop method as example to show here
# reset_index permits us to get output in dataframe rather than a series (in short, it looks prettier)
df_drop = df_drop.groupby(["Artist"])["Likes"].sum().reset_index()

# show top and bottom 10 artists by likes using nsmallest/nlargest
print(
    "DROP METHOD\nUsing nlargest/nsmallest\nTOP 10 ARTISTS BY LIKES\n",
    df_drop.nlargest(n=10, columns=["Likes"]),
)
print("\nBOTTOM 10 ARTISTS BY LIKES\n", df_drop.nsmallest(n=10, columns=["Likes"]))

# using sort_values
df_drop = df_drop.sort_values(by="Likes", ascending=False)

# show top and bottom 10 artists by likes using sort_values
print("\nUsing sort_values\nTOP 10 ARTISTS BY LIKES\n", df_drop.head(10))
print("\nBOTTOM 10 ARTISTS BY LIKES\n", df_drop.tail(10))
```

```python
# create visuals for top 10 and bottom 10 artists by like
# may create individually (easier) or as subplots (harder)

# set subplots
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 5), sharex=False)
# adjust spacing between subplots (if vertical, would use h_pad)
f.tight_layout(w_pad=3)

# create barplots
sns.barplot(data=df_drop.head(10), x="Artist", y="Likes", ax=ax1)
sns.barplot(data=df_drop.tail(10), x="Artist", y="Likes", ax=ax2)

# rotate x-axis tick marks for better visibility
# we cannot use plt.xticks (rotation = []) because of how we have these set up
# were you to show each figure in a different cell, you could use plt.xticks
ax1.tick_params(axis="x", labelrotation=90)
ax2.tick_params(axis="x", labelrotation=90)

# remove notation of 'le8' for numbers on top 10 and instead show full number with comma separator
ax1.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, p: format(int(x), ",")))

# set titles for each visual
ax1.set_title("Top 10 Artists by Likes")
ax2.set_title("Bottom 10 Artists by Likes")

plt.show()
```

##### 2. Find the 10 longest songs and 10 shortest songs (use the Duration_ms column)
Note: The Duration_ms column has null values, so you'll need to decide how to handle those.

```python
# how many null values do you need to handle?
# how does this influence how you'll proceed?

print("Duration_ms column NaN/nulls: ", df["Duration_ms"].isna().sum())
print("Total df records: ", len(df.index))
print(
    "Percent NaN/Null: ",
    round((df["Duration_ms"].isna().sum() / len(df.index)) * 100, 2),
    "%",
)
```

```python
# handle nulls
# you can choose to create a df copy or to edit the df directly
# if you want to play with different ways of handling nulls, create a copy

# average
# df_avg = df.copy() # not needed since this was created earlier (see cell 6), but if you didn't use this method earlier, you'll need this line now
df_avg["Duration_ms"].fillna((df_avg["Duration_ms"].mean()), inplace=True)
print("\nFILLNA AVERAGE")
print(
    "Duration_ms column NaN/Nulls after fillna average: ",
    df_avg["Duration_ms"].isna().sum(),
)
print("Total df records original: ", len(df.index))
print("Total df records after fillna average: ", len(df_avg.index))
```

```python
# get 10 shortest and 10 longest songs

# for simplicity, will keep only relevant columns
# creating copy in case want to use df_avg again
df_avg_trunc = df_avg.copy()
df_avg_trunc = df_avg_trunc.loc[:, ["Artist", "Track", "Duration_ms"]]

# just makes it so the df prints prettier
pd.set_option("display.width", 1000)

# show top and bottom 10 artists by likes using nsmallest/nlargest
print(
    "FILLNA MEAN METHOD\nUsing nlargest/nsmallest\nLONGEST 10 SONGS\n",
    df_avg_trunc.nlargest(n=10, columns=["Duration_ms"]),
)
print("\nSHORTEST 10 SONGS\n", df_avg_trunc.nsmallest(n=10, columns=["Duration_ms"]))

# using sort_values
df_avg = df_avg.sort_values(by="Duration_ms", ascending=False)

# show top and bottom 10 artists by likes using sort_values
print("\nUsing sort_values\nLONGEST 10 SONGS\n", df_avg_trunc.head(10))
print("\nSHORTEST 10 SONGS\n", df_avg_trunc.tail(10))
```

```python
# create visuals for longest 10 and shortest 10 songs
# may create individually (easier) or as subplots (harder)

# set subplots
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 5), sharex=False)

# adjust spacing between subplots (if vertical, would use h_pad)
f.tight_layout(w_pad=3)

# create barplots
sns.barplot(
    data=df_avg_trunc.nlargest(n=10, columns=["Duration_ms"]),
    x="Track",
    y="Duration_ms",
    ax=ax1,
)
sns.barplot(
    data=df_avg_trunc.nsmallest(n=10, columns=["Duration_ms"]),
    x="Track",
    y="Duration_ms",
    ax=ax2,
)

# rotate x-axis tick marks for better visibility
# we cannot use plt.xticks (rotation = []) because of how we have these set up
# were you to show each figure in a different cell, you could use plt.xticks
ax1.tick_params(axis="x", labelrotation=90)
ax2.tick_params(axis="x", labelrotation=90)

# remove notation of 'le8' for numbers on top 10 and instead show full number with comma separator
ax1.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, p: format(int(x), ",")))
ax2.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, p: format(int(x), ",")))

# set titles for each visual
ax1.set_title("Longest 10 Songs")
ax2.set_title("Shortest 10 Songs")

plt.show()
```

Well that looks weird, doesn't it? 
Multiple artists can have the same track name.
To eliminate this as an issue for our visual, we'll create a new column that concatenates track and artist and then recreate the visual.

```python
# create concatenated column
df_avg_trunc["Concat_Track_Artist"] = (
    df_avg_trunc["Track"] + " - " + df_avg_trunc["Artist"]
)
df_avg_trunc
```

```python
# create visual using Duration_ms and Concat_Track_Artist rather than Duration_ms and Track

# set subplots
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 5), sharex=False)

# adjust spacing between subplots (if vertical, would use h_pad)
f.tight_layout(w_pad=3)

# create barplots
sns.barplot(
    data=df_avg_trunc.nlargest(n=10, columns=["Duration_ms"]),
    x="Concat_Track_Artist",
    y="Duration_ms",
    ax=ax1,
)
sns.barplot(
    data=df_avg_trunc.nsmallest(n=10, columns=["Duration_ms"]),
    x="Concat_Track_Artist",
    y="Duration_ms",
    ax=ax2,
)

# rotate x-axis tick marks for better visibility
# we cannot use plt.xticks (rotation = []) because of how we have these set up
# were you to show each figure in a different cell, you could use plt.xticks
ax1.tick_params(axis="x", labelrotation=90)
ax2.tick_params(axis="x", labelrotation=90)

# remove notation of 'le8' for numbers on top 10 and instead show full number with comma separator
ax1.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, p: format(int(x), ",")))
ax2.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, p: format(int(x), ",")))

# set titles for each visual
ax1.set_title("Longest 10 Songs")
ax2.set_title("Shortest 10 Songs")

plt.show()
```

##### 3. Find the top 10 songs by play count (add the Views + Stream columns together)

```python
# how many null values do you need to handle?
# how does this influence how you'll proceed?

print("Views column NaN/nulls: ", df["Views"].isna().sum())
print("Streams column NaN/nulls: ", df["Stream"].isna().sum())
print("Total df records: ", len(df.index))
print(
    "Percent NaN/Null: ",
    round((df["Views"].isna().sum() / len(df.index)) * 100, 2),
    "%",
)
print(
    "Percent NaN/Null: ",
    round((df["Stream"].isna().sum() / len(df.index)) * 100, 2),
    "%",
)
```

```python
# handle nulls
# you can choose to create a df copy or to edit the df directly
# if you want to play with different ways of handling nulls, create a copy

# using median for example

# median
df_med = df.copy()
df_med["Stream"].fillna((df_med["Stream"].median()), inplace=True)
df_med["Views"].fillna((df_med["Views"].median()), inplace=True)

print("\nFILLNA MEDIAN")
print("Stream column NaN/Nulls after fillna median: ", df_med["Stream"].isna().sum())
print("Views column NaN/Nulls after fillna median: ", df_med["Views"].isna().sum())
print("Total df records original: ", len(df.index))
print("Total df records (Stream) after fillna median: ", len(df_med["Stream"].index))
print("Total df records (Views) after fillna median: ", len(df_med["Views"].index))
```

```python
# add views and stream columns together
df_med["Total V&S"] = df_med["Views"] + df_med["Stream"]
df_med.head(3)
```

```python
# get top 10 and bottom 10 by play count

# for simplicity, will keep only relevant columns
# creating copy in case want to use df_avg again
df_med_trunc = df_med.copy()
df_med_trunc = df_med_trunc.loc[:, ["Artist", "Track", "Total V&S"]]

# show top and bottom 10 artists by likes using nsmallest/nlargest
print(
    "FILLNA MEDIAN METHOD\nUsing nlargest/nsmallest\nLONGEST 10 SONGS\n",
    df_med_trunc.nlargest(n=10, columns=["Total V&S"]),
)
print("\nSHORTEST 10 SONGS\n", df_med_trunc.nsmallest(n=10, columns=["Total V&S"]))

# using sort_values
df_med_trunc = df_med_trunc.sort_values(by="Total V&S", ascending=False)

# show top and bottom 10 artists by likes using sort_values

print("\nUsing sort_values\nTOP 10 SONGS BY PLAY COUNT\n", df_med_trunc.head(10))
print("\nBOTTOM 10 SONGS BY PLAY COUNT\n", df_med_trunc.tail(10))
```

```python
# create visual

# set subplots
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 5), sharex=False)

# adjust spacing between subplots (if vertical, would use h_pad)
f.tight_layout(w_pad=3)

# create barplots
sns.barplot(
    data=df_med_trunc.nlargest(n=10, columns=["Total V&S"]),
    x="Track",
    y="Total V&S",
    ax=ax1,
)
sns.barplot(
    data=df_med_trunc.nsmallest(n=10, columns=["Total V&S"]),
    x="Track",
    y="Total V&S",
    ax=ax2,
)

# rotate x-axis tick marks for better visibility
# we cannot use plt.xticks (rotation = []) because of how we have these set up
# were you to show each figure in a different cell, you could use plt.xticks
ax1.tick_params(axis="x", labelrotation=90)
ax2.tick_params(axis="x", labelrotation=90)

# remove notation of 'le8' for numbers on top 10 and instead show full number with comma separator
ax1.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, p: format(int(x), ",")))
ax2.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, p: format(int(x), ",")))

# set titles for each visual
ax1.set_title("Most Played 10 Songs")
ax2.set_title("Least Played 10 Songs")

plt.show()
```

##### Top 10 artists by play count (add the Views + Stream columns together, then sum that result and group by Artist)
Views and Streams both have null values.

```python
# how many null values do you need to handle?
# how does this influence how you'll proceed?

print("Views column NaN/nulls: ", df["Views"].isna().sum())
print("Streams column NaN/nulls: ", df["Stream"].isna().sum())
print("Total df records: ", len(df.index))
print(
    "Percent NaN/Null: ",
    round((df["Views"].isna().sum() / len(df.index)) * 100, 2),
    "%",
)
print(
    "Percent NaN/Null: ",
    round((df["Stream"].isna().sum() / len(df.index)) * 100, 2),
    "%",
)
```

```python
# handle nulls
# you can choose to create a df copy or to edit the df directly
# if you want to play with different ways of handling nulls, create a copy

# using mode for example
df_mod = df.copy()

df_mod["Stream"].fillna((df_mod["Stream"].mode()[0]), inplace=True)
df_mod["Views"].fillna((df_mod["Views"].mode()[0]), inplace=True)

print("\nFILLNA MODE")
print("Stream column NaN/Nulls after fillna mode: ", df_mod["Stream"].isna().sum())
print("Views column NaN/Nulls after fillna mode: ", df_mod["Views"].isna().sum())
print("Total df records original: ", len(df.index))
print("Total df records (Stream) after fillna mode: ", len(df_mod["Stream"].index))
print("Total df records (Views) after fillna mode: ", len(df_mod["Views"].index))
```

```python
# add Views and Stream columns together
df_mod["Total V&S"] = df_mod["Views"] + df_mod["Stream"]
df_mod.head(3)
```

```python
# group likes by artist and get top and bottom 10 artists by play count

# selected drop method as example to show here
# reset_index permits us to get output in dataframe rather than a series (in short, it looks prettier)
df_mod = df_mod.groupby(["Artist"])["Total V&S"].sum().reset_index()

# show top and bottom 10 artists by likes using nsmallest/nlargest
print(
    "MODE METHOD\nUsing nlargest/nsmallest\nTOP 10 ARTISTS BY PLAY COUNT\n",
    df_mod.nlargest(n=10, columns=["Total V&S"]),
)
print(
    "\nBOTTOM 10 ARTISTS BY PLAY COUNT\n", df_mod.nsmallest(n=10, columns=["Total V&S"])
)

# using sort_values
df_mod = df_mod.sort_values(by="Total V&S", ascending=False)

# show top and bottom 10 artists by likes using sort_values
print("\nUsing sort_values\nTOP 10 ARTISTS BY PLAY COUNT\n", df_mod.head(10))
print("\nBOTTOM 10 ARTISTS BY PLAY COUNT\n", df_mod.tail(10))
```

```python
# create visual

# set subplots
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 5), sharex=False)

# adjust spacing between subplots (if vertical, would use h_pad)
f.tight_layout(w_pad=3)

# create barplots
sns.barplot(
    data=df_mod.nlargest(n=10, columns=["Total V&S"]), x="Artist", y="Total V&S", ax=ax1
)
sns.barplot(
    data=df_mod.nsmallest(n=10, columns=["Total V&S"]),
    x="Artist",
    y="Total V&S",
    ax=ax2,
)

# rotate x-axis tick marks for better visibility
# we cannot use plt.xticks (rotation = []) because of how we have these set up
# were you to show each figure in a different cell, you could use plt.xticks
ax1.tick_params(axis="x", labelrotation=90)
ax2.tick_params(axis="x", labelrotation=90)

# remove notation of 'le[x]' for numbers on top 10 and instead show full number with comma separator
# can use this if not scaling to millions or higher as fits your data
# ax1.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, p: format(int(x), ',')))
# ax2.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, p: format(int(x), ',')))

# show ax1 y axis in billions
ax1.yaxis.set_major_formatter(
    mticker.FuncFormatter(lambda x, pos: "{0:g}".format(x / 1e8))
)
ax1.set_ylabel("val in billions")

# show ax 2 y axis in millions
ax2.yaxis.set_major_formatter(
    mticker.FuncFormatter(lambda x, pos: "{0:g}".format(x / 1e6))
)
ax2.set_ylabel("val in millions")

# set titles for each visual
ax1.set_title("10 Most Played Artists")
ax2.set_title("10 Least Played Artists")

plt.show()
```
